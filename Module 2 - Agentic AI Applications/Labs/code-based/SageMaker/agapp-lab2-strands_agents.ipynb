    <h2 style="color: #2f1381;">1. Introduction to AI Agents and the Strands SDK</h2>
</div>

An agent is a sophisticated software construct designed to perceive its surroundings, make informed decisions, execute actions, and learn from experiences in a continuous, iterative cycle. The intelligence stems from the interplay of perception, memory, reasoning, action execution, and feedback loops.

The Strands Agents SDK is an open-source framework that streamlines AI agent development through a "model-driven" approach, leveraging the inherent reasoning capabilities of Large Language Models (LLMs). Key advantages include:

- **Simplicity**: Agents built with minimal code by defining prompts and tools
- **Model Agnosticism**: Support for various LLM providers (Bedrock, Anthropic, OpenAI, etc.)
- **Rich Tooling**: Pre-built tools for common functionalities
- **Production-Ready**: Built-in observability, security, and scalability features

<div id="section2" style="border-left: 5px solid #2f1381; padding-left: 15px; margin: 40px 0 20px 0;">
    <h2 style="color: #2f1381;">2. The Agent Loop: The Brains Behind Autonomous Behavior</h2>
</div>

The agent loop is the fundamental operational cycle that empowers AI agents to exhibit intelligent, autonomous behavior. This continuous process follows a "Think-Act-Learn" paradigm:

1. **Reasoning (Think)**: The agent examines its current goal and context, then plans its next course of action
2. **Action (Act)**: The agent executes the decided action, typically by invoking external tools
3. **Reflection (Learn)**: The agent evaluates results and adjusts its understanding for future plans

Within the Strands SDK, this loop is implemented as a model-guided cycle:

1. **User Input Processing**: Adding messages to conversation history
2. **Model Processing**: LLM receives system prompt, history, and tool configurations
3. **Response Analysis & Tool Execution**: Extract, validate, and execute tool calls if needed
4. **Tool Result Processing**: Format results and add back to conversation history
5. **Recursive Processing**: Continue loop if further actions needed
6. **Completion**: Generate final response or handle exceptions

<img src="images/Agent Loop _ Strands Agents SDK.png" alt="Activity" width="800"/>

Let's see this in action with a simple iterative example:

```python
from strands import Agent
from strands_tools import calculator
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

# Initialize the agent with calculator tool and system prompt
agent = Agent(
    model=bedrock_model,
    tools=[calculator],
    system_prompt="You are a helpful mathematical assistant. Always break down complex calculations into steps if needed."
)

print("Agent initialized. Ready to calculate!")

# First interaction: simple calculation
first_prompt = "What is 123456789 + 987654321?"
print(f"\nUser: {first_prompt}")
response1 = agent(first_prompt)
print(f"\nAgent: {response1.message}")

# Second interaction: builds on previous result using conversation history
second_prompt = "Now divide the result by 23."
print(f"\nUser: {second_prompt}")
response2 = agent(second_prompt)
print(f"\nAgent: {response2.message}")
```

<div id="section3-1" style="border-left: 3px solid #4C32E2; padding-left: 12px; margin: 25px 0 15px 15px;">
    <h3 style="color: #4C32E2;">3.1 Conversation Management in Strands</h3>
</div>

Memory is fundamental for intelligent agents, enabling context maintenance across interactions. Without memory, agents would treat every input as isolated, leading to disjointed experiences.

**Types of Memory:**
- **Working Memory**: Short-term context during active sessions, managed through `agent.messages`
- **Persistent Memory**: Long-term recall beyond single sessions, using external storage

Since LLMs operate within token limits, Strands addresses "token buildup" through Conversation Managers that intelligently curate conversation history.

The default `SlidingWindowConversationManager` maintains only the most recent N message pairs, automatically discarding older messages when the window size is exceeded:

```python
from strands import Agent
from strands.agent.conversation_manager import SlidingWindowConversationManager
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

# Create conversation manager with custom window size
conversation_manager = SlidingWindowConversationManager(window_size=2)

# Initialize agent with custom conversation manager
agent = Agent(
    model = bedrock_model,
    conversation_manager=conversation_manager,
    system_prompt="You are like Dory from Finding Nemo. You might forget some interactions we have had, and that is alright."
)

print("Agent initialized with a sliding window of 2 messages.")
print("-" * 80)

# Define specific messages to demonstrate memory limitations
messages = [
    "My favorite color is blue. Remember this for later.",
    "I'm planning a trip to Hawaii next month.",
    "I think artificial intelligence is fascinating to learn about.",
    "The weather today is quite nice, don't you think?",
    "I am learning how to play the piano.",
    "What did I say was my favorite color in our first conversation?"
]

# Process each message and observe the sliding window effect
for i, message in enumerate(messages, 1):
    print(f"\n📝 User message {i}: \"{message}\"")
    response = agent(message)
    print(f"\n🤖 Agent: \"{response.message['content'][0]['text']}\"")
    print(f"\n📊 Messages in history: {len(agent.messages)}")
    print("-" * 80)

# Highlight the final response to emphasize memory limitation
print("\n🔍 OBSERVATION:")
print("The agent should be unable to recall the favorite color (blue) from the first message")
print("because it was pushed out of the sliding window of 2 messages.")
print("-" * 80)
print("Final response to color question:")
print(f"\n🤖 \"{response.message['content'][0]['text']}\"")
```

<div id="section3-2" style="border-left: 3px solid #4C32E2; padding-left: 12px; margin: 25px 0 15px 15px;">
    <h3 style="color: #4C32E2;">3.2 Session Persistence with FileSessionManager</h3>
</div>

For applications requiring context across separate interactions or distributed deployments, session persistence is crucial. Strands offers session management abstractions to automatically persist and restore agent conversations.

The `FileSessionManager` is ideal for local development, storing session data directly in files on the filesystem. Session persistence is automatically triggered by agent initialization, message addition, invocation, and redaction events.

Let's demonstrate saving and loading agent conversation history:

```python
import os
import shutil
from strands import Agent
from strands.session.file_session_manager import FileSessionManager
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

# Define directory and session ID
SESSION_DIR = "agent_sessions"
os.makedirs(SESSION_DIR, exist_ok=True)
session_id = "my-unique-session-123"

print(f"📁 Session data for '{session_id}' will be stored in: {SESSION_DIR}")
print("=" * 80)

# First Run: Create and interact with agent
print("\n🔵 FIRST RUN - Creating New Agent Session")
print("-" * 80)
session_manager_1 = FileSessionManager(session_id=session_id, storage_dir=SESSION_DIR)

agent1 = Agent(
    model=bedrock_model,
    session_manager=session_manager_1,
    system_prompt="You are a helpful assistant."
)
print("✅ Agent 1 initialized with new session")

# First interaction
print("\n👤 User: What are AI Agents?")
response1 = agent1("What are AI Agents?")
print(f"\n🤖 Agent: {response1.message['content'][0]['text']}")

# Second interaction
print("\n👤 User: Can you tell me a fun fact about space?")
response2 = agent1("Can you tell me a fun fact about space?")
print(f"\n🤖 Agent: {response2.message['content'][0]['text']}")

print(f"\n📊 Agent 1 conversation history: {len(agent1.messages)} messages")
print(f"💾 Session saved to: {SESSION_DIR}")
print("=" * 80)

# Second Run: Simulate new application run and restore agent
print("\n🔴 SECOND RUN - Restoring Previous Agent Session")
print("-" * 80)
session_manager_2 = FileSessionManager(session_id=session_id, storage_dir=SESSION_DIR)

agent2 = Agent(
    model=bedrock_model,
    session_manager=session_manager_2,
    system_prompt="You are a helpful assistant."
)
print("✅ Agent 2 initialized (restored from previous session)")

# Test memory retention
print("\n👤 User: What was the last thing we talked about?")
response3 = agent2("What was the last thing we talked about?")
print(f"\n🤖 Agent: {response3.message['content'][0]['text']}")

print(f"\n📊 Agent 2 conversation history: {len(agent2.messages)} messages")
print("=" * 80)

# Clean up
shutil.rmtree(SESSION_DIR)
print(f"\n🧹 Cleaned up session: {SESSION_DIR}")
```

<div id="section5" style="border-left: 5px solid #2f1381; padding-left: 15px; margin: 40px 0 20px 0;">
    <h2 style="color: #2f1381;">5. Tool Calling: Empowering Agents to Act</h2>
</div>

Tool calling refers to the ability of AI models to interact with external tools, APIs, or systems to enhance their functionalities. This capability transforms agents from text generators into action-taking systems that can access real-time information, perform precise calculations, and interact with external databases.

**How Strands Agents Utilize Tools:**

1. **Recognizing the Need**: LLM identifies that requested information cannot be fulfilled from internal knowledge
2. **Selecting Tools**: Based on context and tool descriptions, LLM selects appropriate tools
3. **Constructing Queries**: LLM formulates structured requests with necessary arguments
4. **Processing Results**: Tool output is captured, formatted, and fed back into LLM context

**Overview of strands_tools Package:**

The strands-agents-tools package provides pre-built tools that significantly extend agent capabilities:

| Tool Name | Use Case | Notes |
|-----------|----------|-------|
| calculator | Mathematical operations, symbolic math, equation solving | Widely used for precise calculations |
| current_time | Getting current date/time in specified timezones | Useful for time-sensitive tasks |
| file_read | Reading configuration files, parsing code, loading datasets | Essential for file system interaction |
| file_write | Writing results, creating files, saving output data | Complements file_read for complete file management |
| python_repl | Running Python code, data analysis, complex logic execution | Powerful for dynamic scripting; requires user confirmation |
| generate_image | Creating AI-generated images from text descriptions | Leverages image generation models |
| nova_reels | Creating high-quality videos using Amazon Bedrock Nova Reel | Specialized tool for video generation |
| image_reader | Processing and analyzing image files | Enables multimodal input capabilities |

Let's explore these tools with practical examples:

<!-- Subsection Header -->
<div id="section5-1" style="border-left: 3px solid #4C32E2; padding-left: 12px; margin: 25px 0 15px 15px;">
    <h3 style="color: #4C32E2;">5.1 File Operations (file_read, file_write)</h3>
</div>

File operation tools enable agents to interact with the local file system, crucial for data processing, configuration management, and report generation:

```python
import os
from strands import Agent
from strands_tools import file_read, file_write
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

print("📁 FILE OPERATIONS AGENT DEMONSTRATION")
print("=" * 80)

# Enable bypass for tool consent to avoid interactive prompts
os.environ["BYPASS_TOOL_CONSENT"] = "true"
print("✅ Tool consent bypass enabled for automated execution")

# Define file path and content for demonstration
file_path = "agent_output.txt"
content_to_write = "This is a test message written by a Strands agent.\nIt demonstrates file_write capabilities."

# Initialize agent with file operation tools
agent = Agent(
    model=bedrock_model,
    tools=[file_write, file_read],
    system_prompt="You are a file management assistant. You can write to and read from files."
)

print("✅ Agent initialized with file_read and file_write tools")
print("-" * 80)

# Write content to file
print(f"\n👤 User: Write the following content to '{file_path}':")
print(f"📝 Content: \"{content_to_write}\"")

response_write = agent(f"Write the following content to a file named '{file_path}': {content_to_write}")
try:
    write_response_text = response_write.message['content'][0]['text']
    print(f"\n🤖 Agent: {write_response_text}")
except (KeyError, IndexError):
    print(f"\n🤖 Agent: {response_write.message}")

# Verify file creation
if os.path.exists(file_path):
    print(f"✅ File '{file_path}' successfully created")
else:
    print(f"❌ File creation failed for '{file_path}'")
print("-" * 80)

# Read content from file
print(f"\n👤 User: Now, read the content of the file '{file_path}'.")

response_read = agent(f"Read the content of the file '{file_path}'.")
try:
    read_response_text = response_read.message['content'][0]['text']
    print(f"\n🤖 Agent: {read_response_text}")
except (KeyError, IndexError):
    print(f"\n🤖 Agent: {response_read.message}")
print("-" * 80)

# Clean up
if os.path.exists(file_path):
    os.remove(file_path)
    print(f"🧹 Cleaned up file: {file_path}")
print("=" * 80)
```

<div id="section5-2" style="border-left: 3px solid #4C32E2; padding-left: 12px; margin: 25px 0 15px 15px;">
    <h3 style="color: #4C32E2;">5.2 Python REPL (python_repl)</h3>
</div>

    <h3 style="color: #4C32E2;">5.3 Image Generation (generate_image)</h3>
</div>

The generate_image tool empowers agents to create AI-generated images from textual prompts. This opens creative applications for visualizing concepts, generating artistic content, or producing visual aids.

```python
from strands import Agent, tool
import base64
import json
import os
import random
import boto3
from typing import Optional

@tool
def generate_image_nova(prompt: str, path: str = 'generated_images/sample.png', width: int = 512, height: int = 512, quality: str = "standard") -> str:
    """Generate an image using Amazon Nova Canvas based on the provided text prompt.
    
    Args:
        prompt: Text description of the image to generate
        path: Path to store the images
        width: Image width in pixels (default: 512)
        height: Image height in pixels (default: 512)
        quality: Image quality setting, can be "standard" or "premium" (default: "standard")
    
    Returns:
        Path to the generated image file
    """
    # Create a Bedrock Runtime client
    client = boto3.client("bedrock-runtime", region_name="us-east-1")
    
    # Set the model ID
    model_id = "amazon.nova-canvas-v1:0"
    
    # Generate a random seed
    seed = random.randint(0, 858993460)
    
    # Format the request payload
    native_request = {
        "taskType": "TEXT_IMAGE",
        "textToImageParams": {"text": prompt},
        "imageGenerationConfig": {
            "seed": seed,
            "quality": quality,
            "height": height,
            "width": width,
            "numberOfImages": 1,
        },
    }
    
    # Convert the request to JSON
    request = json.dumps(native_request)
    
    # Invoke the model
    response = client.invoke_model(modelId=model_id, body=request)
    
    # Decode the response
    model_response = json.loads(response["body"].read())
    
    # Extract the image data
    base64_image_data = model_response["images"][0]
    
    # Save the image to a local folder
    if not os.path.exists(os.path.dirname(path)):
        os.makedirs(os.path.dirname(path))
   
    # Save the image
    image_data = base64.b64decode(base64_image_data)
    image_path = path
    
    with open(image_path, "wb") as file:
        file.write(image_data)
    
    return image_path
```

```python
import os
from strands import Agent
from IPython.display import Image, display
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

print("🎨 IMAGE GENERATION AGENT DEMONSTRATION")
print("=" * 80)

# Initialize agent with generate_image tool
agent = Agent(
    model=bedrock_model,
    tools=[generate_image_nova],
    system_prompt="You are a creative image generation assistant. You can create images from descriptions"
)

print("✅ Agent initialized with generate_image tool")

# Define output directory
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)
print(f"📁 Images will be saved to: {output_dir}")
print("-" * 80)

# Generate landscape image
prompt1 = "A serene landscape with a river flowing through it, in a watercolor style."
output_path1 = os.path.join(output_dir, "serene_landscape.png")
print(f"\n👤 User: Generate an image: '{prompt1}' and save it to '{output_path1}'")

response1 = agent(f"Generate an image of '{prompt1}' and save it as '{output_path1}'.")
try:
    response1_text = response1.message['content'][0]['text']
    print(f"\n🤖 Agent: {response1_text}")
except (KeyError, IndexError):
    print(f"\n🤖 Agent: {response1.message}")

# Check if image was created
if os.path.exists(output_path1):
    print(f"✅ Image successfully generated: {output_path1}")
    # Display the image in the notebook
    display(Image(filename=output_path1, width=400, height=400))
else:
    print(f"❌ Image generation failed for: {output_path1}")
print("-" * 80)

# Generate futuristic city image
prompt2 = "A futuristic city skyline at sunset, cyberpunk aesthetic."
output_path2 = os.path.join(output_dir, "cyberpunk_city.png")
print(f"\n👤 User: Generate an image: '{prompt2}' and save it to '{output_path2}'")

response2 = agent(f"Generate an image of '{prompt2}' and save it as '{output_path2}'.")
try:
    response2_text = response2.message['content'][0]['text']
    print(f"\n🤖 Agent: {response2_text}")
except (KeyError, IndexError):
    print(f"\n🤖 Agent: {response2.message}")

# Check if image was created
if os.path.exists(output_path2):
    print(f"✅ Image successfully generated: {output_path2}")
    display(Image(filename=output_path2, width=400, height=400))
else:
    print(f"❌ Image generation failed for: {output_path2}")
print("-" * 80)

# Cleanup code to delete the generated_images directory
print("\n🧹 Cleaning up generated images...")
try:
    shutil.rmtree(output_dir)
    print(f"✅ Successfully deleted the '{output_dir}' directory")
except Exception as e:
    print(f"❌ Failed to delete the '{output_dir}' directory: {e}")
```

**Problem**: Users encounter various issues with smart devices, from connectivity problems to software glitches. Traditional troubleshooting involves complex navigation or extensive knowledge base searches.

**Agent Design**: An intelligent troubleshooting assistant needs:
- **Persona**: Patient, knowledgeable technical support character
- **Multimodal Input**: Process text descriptions and error images via `image_reader`
- **Diagnostic Capabilities**: Use `file_read` for logs and `python_repl` for simulated diagnostics
- **Session Management**: Maintain context across multiple troubleshooting turns

Let's implement this comprehensive use case:

```python
import os
import shutil
from strands import Agent
from strands_tools import image_reader, file_read, python_repl
from strands.session.file_session_manager import FileSessionManager
from mlu_utils.lab2_utils.helper import get_bedrock_model

bedrock_model = get_bedrock_model("amazon.nova-pro-v1:0")

print("🔧 ECHO DEVICE TROUBLESHOOTING ASSISTANT DEMONSTRATION")
print("=" * 80)

# Set environment variable to bypass tool consent
os.environ["BYPASS_TOOL_CONSENT"] = "true"
if os.path.exists('repl_state'):
    shutil.rmtree('repl_state')
if os.path.exists("troubleshooting_sessions"):
    shutil.rmtree('troubleshooting_sessions')


# Use real troubleshooting resources
docs_dir = "mlu_utils/data/docs"

# Setup session management
SESSION_DIR = "troubleshooting_sessions"
os.makedirs(SESSION_DIR, exist_ok=True)
session_id = "echo-troubleshoot-user-xyz"
session_manager = FileSessionManager(session_id=session_id, storage_dir=SESSION_DIR)
print(f"💾 Session manager configured for session ID: {session_id}")

# Initialize troubleshooting agent
troubleshooting_system_prompt = """You are an expert Amazon Echo device troubleshooting assistant.
Your goal is to help users diagnose and resolve issues with their Echo devices using the documents provided.
You have access to troubleshooting documentation at "mlu_utils/data/docs".

IMPORTANT INSTRUCTIONS FOR FILE SEARCHING:
1. First, use file_read with action='tree' to list all available files in the directory
2. When searching for information:
   - If exact pattern matching returns 0 results, DO NOT stop there
   - Instead, use file_read with action='read' to read the FULL CONTENT of the most relevant file
   - Look for files with names related to the user's issue (e.g., "wifi" for Wi-Fi issues)
3. Always provide helpful responses even if exact matches aren't found
4. Be patient, empathetic, and guide the user step-by-step
5. If you need more information, ask specific questions

WORKFLOW:
- Step 1: List files with action='tree'
- Step 2: Identify relevant file based on filename
- Step 3: Read full file content with action='read' and path to the specific file
- Step 4: Extract relevant information and provide guidance"""

agent = Agent(
    model=bedrock_model,
    tools=[image_reader, file_read, python_repl],
    system_prompt=troubleshooting_system_prompt,
    session_manager=session_manager
)
print("✅ Echo Troubleshooting Agent initialized with multimodal capabilities")
print("-" * 80)

# Step 1: User describes problem
print("\n👤 User: My Echo device isn't connecting to Wi-Fi.")
response1 = agent("My Echo device isn't connecting to Wi-Fi.")
try:
    response1_text = response1.message['content'][0]['text']
    print(f"\n🤖 Agent: {response1_text}")
except (KeyError, IndexError):
    print(f"\n🤖 Agent: {response1.message}")
print("-" * 80)
```
