First, let's configure our AWS clients for Bedrock:

```python
# Configure AWS bedrock_client
bedrock_client = boto3.client('bedrock')
bedrock_runtime = boto3.client('bedrock-runtime')
```

Now, let's create a guardrail that prevents providing medical advice, restricts content types, and anonymizes sensitive data:

```python
import time

#personally identifiable information types
pii_types = [
    
    "AWS_ACCESS_KEY",
    "AWS_SECRET_KEY",
    "CA_HEALTH_NUMBER",
    "CA_SOCIAL_INSURANCE_NUMBER",
    "CREDIT_DEBIT_CARD_CVV",
    "CREDIT_DEBIT_CARD_EXPIRY",
    "CREDIT_DEBIT_CARD_NUMBER",
    "NAME",
    "PASSWORD",
    "UK_NATIONAL_HEALTH_SERVICE_NUMBER",
    "UK_NATIONAL_INSURANCE_NUMBER",
    "US_BANK_ACCOUNT_NUMBER",
    "US_BANK_ROUTING_NUMBER",
    "US_SOCIAL_SECURITY_NUMBER"
]

guardrail_name_prefix = 'no-medical-advice'

timestamp = int(time.time())
guardrail_name = f"{guardrail_name_prefix}_{timestamp}"

# Create a Bedrock guardrail
response = bedrock_client.create_guardrail(
    name=guardrail_name,
    description='Prevents the model from providing medical advice.',
    topicPolicyConfig={
        'topicsConfig': [
            {
                'name': 'Medical Advice',
                'definition': 'Providing personalized medical advice, diagnosis, treatment recommendations, or prescriptions that should only be given by qualified healthcare professionals.',
                'examples': [
                    'What medication should I take for my symptoms?',
                    'Do I need surgery for this condition?',
                    'How should I adjust my insulin dosage?',
                    'Is this mole cancerous?',
                    'Should I stop taking this prescription medication?'
                ],
                'type': 'DENY'
            }
        ]
    },
    contentPolicyConfig={
        'filtersConfig': [
            {
                'type': 'SEXUAL',
                'inputStrength': 'HIGH', # you can imagine a case where a medical term may trigger this guardrail. use with caution and testing
                'outputStrength': 'HIGH'
            },
            {
                'type': 'VIOLENCE',
                'inputStrength': 'HIGH', # you can imagine a case where a medical scenario may trigger this guardrail. use with caution and testing
                'outputStrength': 'HIGH'
            },
            {
                'type': 'HATE',
                'inputStrength': 'HIGH',# you can imagine a case where medically relevant demographics or descriptions may trigger this guardrail. use with caution and testing
                'outputStrength': 'HIGH'
            },
            {
                'type': 'INSULTS',
                'inputStrength': 'HIGH',
                'outputStrength': 'HIGH'
            },
            {
                'type': 'MISCONDUCT',
                'inputStrength': 'HIGH',
                'outputStrength': 'HIGH'
            },
            {
                'type': 'PROMPT_ATTACK',
                'inputStrength': 'HIGH',
                'outputStrength': 'NONE'
            }
        ]
    },
     sensitiveInformationPolicyConfig={
        "piiEntitiesConfig": list(
            map(lambda x: {"type": x, "action": "ANONYMIZE"}, pii_types)
        ),
    },
    wordPolicyConfig={
        'wordsConfig': [
            {'text': 'medical advice'},
            {'text': 'treatment recommendations'},
            {'text': 'diagnosis'},
            {'text': 'prescription medication'},
            {'text': 'dosage adjustment'},
            {'text': 'medical treatment'},
            {'text': 'clinical recommendation'},
            {'text': 'medical opinion'},
            {'text': 'therapeutic advice'},
            {'text': 'healthcare recommendation'}
        ],
        'managedWordListsConfig': [
            {
                'type': 'PROFANITY'
            }
        ]
    },
    blockedInputMessaging='I apologize, but I am not able to provide medical advice. For your health and safety, please consult with a qualified healthcare professional. You may also be receiving this message if you have accidentally included sensitive information or restricted topics.',
    blockedOutputsMessaging='I apologize, but I am not able to provide medical advice. For your health and safety, please consult with a qualified healthcare professional. You may also be receiving this message if you have accidentally included sensitive information or restricted topics.',
)

# Print the response to get the guardrail ID
print("Guardrail ID:", response.get('guardrailId'))
print("Guardrail ARN:", response.get('guardrailArn'))
```

```python
guardrail_id = response.get('guardrailId')
guardrail_version = "DRAFT"  # Initial version is always DRAFT
```

Let's create a function to test our guardrail directly:

```python
def test_guardrail(text, source_type='INPUT'):
    response = bedrock_runtime.apply_guardrail(
        guardrailIdentifier=guardrail_id,
        guardrailVersion=guardrail_version,
        source=source_type,  # can be 'INPUT' or 'OUTPUT'
        content=[{"text": {"text": text}}]
      )

      
    print(f"Action 🏁: {response.get('action')}")
    print(f"Action Reason 📝: {response.get('actionReason', 'None')}")

    # Check if content was blocked
    is_blocked = response.get('action') == 'GUARDRAIL_INTERVENED'
    if not is_blocked:
        print(f"✅ Content {source_type} NOT blocked")
        print("Responding...")

        model_response = bedrock_runtime.invoke_model(
            modelId="amazon.nova-pro-v1:0",
            body=json.dumps({
                "schemaVersion": "messages-v1",
                "messages": [
                    {
                        "role": "user",
                        "content": [
                            {
                                "text": text + " and keep the response to 3 sentences or less"
                            }
                        ]
                    }
                ],
                "inferenceConfig": {
                    "maxTokens": 2000,
                    "topP": 0.9,
                    "topK": 100,
                    "temperature": 0.0
                }
            })
        )
        # Parse and print the model's response
        response_body = json.loads(model_response['body'].read())
        print(response_body['output']['message']['content'][0]['text'])
      
          

    if is_blocked:
        # Print topic policies that were triggered
        print(f"🦺 GUARDRAIL INTERVENED 🦺 \n⚠️ Content {source_type} blocked: {is_blocked} ")
        assessments = response.get('assessments', [])
        if assessments and 'topicPolicy' in assessments[0]:
            print("❌ Blocked topics:", [topic.get('name') for topic in
  assessments[0]['topicPolicy'].get('topics', []) if topic.get('action') == 'BLOCKED'])

        # Print the modified output if available
        if 'outputs' in response and response['outputs']:
            print("Modified content:", response['outputs'][0].get('text', 'None'))
          

        return response
```

Now let's integrate our guardrail with a Strands Agent. The Strands Agent has access to the following tools (which are located in <code>mlu_utils</code>).

| tool  |  description |
| -------- | ------- |
| `get_patient_profile`   |  looks up patient in `patient_profiles.json` data  |
| `get_patient_lab_results`   | retrieves the medical lab results for the patient   |
| `update_patient_profile`   | edits the patient profile   |

```python
SYSTEM_PROMPT = "You are a helpful assistant that provides information about patient lab results and health data. You do NOT provide medical advice, diagnosis, or suggestions. However, please comment on possible typos or data entry errors."
```

```python
# Create a Bedrock model with guardrail configuration
boto_config = BotocoreConfig(
    retries={"max_attempts": 3, "mode": "standard"},
    connect_timeout=5,
    read_timeout=60
)

bedrock_model = BedrockModel(
    model_id="amazon.nova-pro-v1:0",
    guardrail_id=guardrail_id, #the no-medical-advice guardrail we created above
    guardrail_version=guardrail_version,
    guardrail_redact_output = True,        
    guardrail_redact_input = True  ,
    # Enable trace info for debugging
    guardrail_trace="enabled",
    boto_client_config=boto_config
)

# Create agent with the guardrail-protected model
agent = Agent(
        system_prompt=SYSTEM_PROMPT,
        model=bedrock_model,
        tools=[
            get_patient_profile,
            get_patient_lab_results,
            update_patient_profile
        ]
    )
```

<div id="section4-2" style="border-left: 3px solid #4C32E2; padding-left: 12px; margin: 25px 0 15px 15px;">
    <h3 style="color: #4C32E2;">4.2 Testing the Agent + Guardrail</h3>
</div>

Let's create a helper function to test our agent with the guardrail:

```python
# Helper function to test the agent and check for guardrail interventions
def test_agent_with_guardrail(agent, prompt):
    print(f"\nUser: {prompt}")
    
    # Get agent response
    response = agent(prompt)
    
    # Check for guardrail intervention
    if hasattr(response, 'stop_reason') and response.stop_reason == "guardrail_intervened":
        print("\n ⚠️ GUARDRAIL INTERVENED!")
        
    else:
        return
        
```

```python
def reset_agent():
    agent = Agent(
        system_prompt=SYSTEM_PROMPT,
        model=bedrock_model,
        tools=[
            get_patient_profile,
            get_patient_lab_results,
            update_patient_profile
        ]
    )
    return agent
```

```python
agent = reset_agent()
```

```python
# Test with a safe question that should call the tool get_patient_lab_results and return the results and a short interpretation
print("Testing ✅ safe input:") 
test_agent_with_guardrail(agent, "my patient ID is PAT1")
```
